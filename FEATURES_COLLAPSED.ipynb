Synthetic dataset to test feature collapse: 30 samples and 3 features (grass background, shape, colour). Grass perfectly predicts the cow label, intentionally introducing a shortcut.
---



```python
# import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import torch
import torch.nn as nn
from torch import sigmoid




```

```python
from pandas.io.api import read_csv

print("="*50)
print("Dataset Analysis")
print("="*50)


# Import the Dataset
df = pd.read_excel('cow_dataset2.xlsx')
print("Dataset:")
print(df.to_string(index=False))


# Extract features and labels
X = df[['Grass_Background', 'Shape', 'Colour']].values
y = df['Label'].values
print("\nDataset (first 5 rows):")
print(df.head().to_string(index=False))
print(f"Total samples: {len(y)}")
print(f"Cows (label=1): {sum(y==1)}")
print(f"Non-Cows (label=0): {len(y) - sum(y)}")
print(f"\nLabels y: {y}")


#Convert to tensor
X_tensor = torch.tensor(df[['Grass_Background', 'Shape', 'Colour']].values, dtype=torch.float32)
y_tensor = torch.tensor(df['Label'].values, dtype=torch.float32).reshape(-1, 1)
print(f"nX_tensor shape: {X_tensor.shape}")
print(f"y_tensor shape: {y_tensor.shape}")

print(f"\nFirst three rows of X_tensor")
print(X_tensor[:3])
print(f"\nFirst three rows of y_tensor")
print(y_tensor[:3])

```

Creating a class (three input layers, 10 hidden layer, one output layer)




```python
## creating a class for simple FF-NN (three input layers, 10 hidden layer, one output layer)

class Net(nn.Module):
    def __init__(self, N_in, N_h, N_out):
        super(Net, self).__init__()
        self.linear1 = nn.Linear(N_in, N_h)
        self.linear2 = nn.Linear(N_h, N_out)
    def forward(self, x):
        oh = sigmoid(self.linear1(x))
        oo = sigmoid(self.linear2(oh))
        return oo


```

Creating actual network with three input neuron, ten hidden neurons, one output neuron


```python
# Creating a network with three input neuron, 10 hidden neurons, one output neuron
my_network = Net(3, 10, 1)
print(my_network)

# parameters can be seen in different ways

print(list(my_network.parameters()))

for p in list(my_network.parameters()):
    print(p)

my_network.state_dict()

for p, v in my_network.state_dict().items():
    print(p, v)
```

Creating input and output of the network forward pass

```python
# Creating input
x1 = X_tensor
print(x1)
# Creating output of the network (forward pass)
y1 = my_network(x1)
print(y1)
```

Loss Function

```python
# Loss function

error = nn.MSELoss()

y_compare = y_tensor
loss = error(y1, y_compare)
print(loss)
print(y_compare)
print(f"Loss: {loss.item()}")
```

Forward and Backward Pass

```python
# Forward pass
y1 = my_network(x1)

# Calculate loss
loss = error(y1, y_compare)

# Backward pass
loss.backward()
```

Weights Update

```python
# Weights Update

learning_rate = 0.07
for f in my_network.parameters():
    f.data.sub_(f.grad.data * learning_rate)
```

Checking the loss after the first update

```python
#Checking the loss
y1_new = my_network(x1)
loss_new = error(y1, y_compare)

print(f"Lossafter the firts update: {loss_new.item()}")
```


Training the loop


```python
#Training the loop
my_network.zero_grad()
for epoch in range(2000):
  y_prediction = my_network(x1)
  loss = error(y_prediction, y_compare)
  my_network.zero_grad()
  loss.backward()
  for p in my_network.parameters():
    p.data = p.data - p.grad.data * learning_rate
  if epoch % 1000 == 0:
      print(f"Epoch: {epoch}, Loss: {loss.item():.6f}")


print(f"Loss after training: {loss.item():.6f}")
```

Feature Collapse`. After training a shallow network (one hidden layer and 10 neurons) I evaluated the rank of the hidden activations. A high rank suggests the use of many independent features, while low rank (collapsed) reflects the network's reliance on few features. The rank collapsed from 10 to 2, indicating the model used grass as a shortcut and discarded shape and colour features.



```python
with torch.no_grad():
  H = sigmoid(my_network.linear1(X_tensor)).numpy()


rank_input = np.linalg.matrix_rank(X_tensor.numpy())
rank_hidden = np.linalg.matrix_rank(H)

print("="*50)
print("Feature Collapse Analysis")
print("="*50)
print(f"\nInput tensor X_tensor shape: {X_tensor.shape}")
print(f"Hidden activations H shape: {H.shape}")
print(f"\nRank  of Input: {rank_input}")
print(f"Rank  of Hidden Layer: {rank_hidden} (Hidden Layer max = 10)")


if rank_hidden < 10:
  print(f"Feature Collapse Detected: from 10 to {rank_hidden}")
print(f"The network only learned the feature shortcut (Grass_Background)")
```
